```python
with open('test.txt','a+') as fh:
  s="Today i will be showing... my video games \n"
  fh.write(s)
  b=fh.readlines()
  print(b)
  fh.close()
```

```python
with open('test.txt','a') as fh:
  s="Today i will be showing... my video games \n"
  fh.write(s)
  print(fh)
```

```python
with open('test.txt','w+') as fp:
  s="hello there adil here \n"
  fp.write(s)
  v=fp.readlines()
  print(v)
```
